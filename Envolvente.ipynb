```python
import numpy as np
import matplotlib.pyplot as plt
import obspy
```

```python
#vamos a cargar los archivos
root_folder = "Z:/DAS/2021-11-30_Medidas"

#list all .mat files in the folder
import glob
files = glob.glob(root_folder + "/*.mat")
print(files)

#load all files
import scipy.io
data = [scipy.io.loadmat(file) for file in files]
```

```python
#concat al Fp
Fp = np.concatenate([d['Fp'] for d in data])
```

```python
#plot all axis 0 (time)
plt.plot(np.transpose(Fp))
plt.show()




```

```python
#compute square mean
Fp_mean = np.mean(Fp**2, axis=0)

#plot the mean
plt.plot(Fp_mean)
```

```python
#compute the inverse of the mean
Fp_mean_inv = 1/Fp_mean
#plot
plt.plot(Fp_mean_inv)
```

```python
#try to compute the envelope of Fp_mean_inv
from scipy.signal import hilbert
Fp_mean_inv_envelope = np.abs(hilbert(Fp_mean))
#plot
plt.plot(Fp_mean_inv_envelope)






```
