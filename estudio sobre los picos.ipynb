```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
import daslib
```

```python


#cargamos la grid de los picos
grid = np.load('grid.npy')
len(grid)
```

```python
#elegimo
index = 32
mat = grid[index]
mat.shape
```

```python
#plot mat
plt.imshow(mat, cmap='gray')
```

```python
#plot all rows of mat as lines
for i in range(mat.shape[0]):
    plt.plot(mat[i,:])
plt.show()
```

```python
#compute all peaks using scipy.signal.find_peaks
from scipy.signal import find_peaks
peaks = []
for i in range(mat.shape[0]):
    peaks.append(find_peaks(mat[i,:], height=0.025)[0])

#plot all peaks and signals
for i in range(mat.shape[0]): 
    print(len(peaks[i]))
    plt.plot(mat[i,:])
    plt.plot(peaks[i], mat[i,peaks[i]], "x")
plt.show()

    





```

```python
#make a mat with the peaks as 1
mat_peaks = np.zeros(mat.shape)
for i in range(mat.shape[0]):
    mat_peaks[i, peaks[i]] = 1

#plot mat_peaks
plt.imshow(mat_peaks, cmap='gray')
plt.show()

```

```python
#use hough transform to find the lines
lines = cv2.HoughLines(mat_peaks.astype(np.uint8), 1, np.pi/180, 200)
lines.shape
```
